Logistic regression with ROC curve and ROC AUC score 

```python
# roc curve and aur  
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from matplotlib import pyplot as plt
```

```python
# Generate a data set

X, y = make_classification(n_samples=1000, n_features=2, n_informative=2, n_redundant=0, random_state=1)
```

```python
# tain test split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
# create a default dummy model as 0 as output (majority class)
dummy_model = [0 for _ in range(len(y_test))]
dummy_model
```

```python
# create a basic logistic model 

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
# prediction based on probablity
model_prob = model.predict_proba(X_test)
```

```python
model_prob
```

```python
# focus on positive values 
model_prob = model_prob[:, 1]
model_prob
```

```python
# calculate roc curve
model_auc = roc_auc_score(y_test , model_prob)
dummy_model_auc = roc_auc_score(y_test , dummy_model)
print("dummy model auc " + str(dummy_model_auc))
print("model auc " + str(model_auc))
```

```python
# if we plot TPR and FPR in x and y coordinates then we get ROC
dummy_fpr_ , dummy_tpr_ = roc_curve(y_test , dummy_model)
model_fpr , model_tpr_  = roc_curve(y_test , model_prob)
```
