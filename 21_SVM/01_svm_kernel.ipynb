```python
# add a space dimension 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x = np.linspace(-5.0, 5.0, 100)
y = np.sqrt(10**2 - x**2)
y=np.hstack([y,-y])
x=np.hstack([x,-x])
```

```python
y
```

```python
x1 = np.linspace(-5.0, 5.0, 100)
y1 = np.sqrt(5**2 - x1**2)
y1=np.hstack([y1,-y1])
x1=np.hstack([x1,-x1])
```

```python
# outer circle
plt.scatter(y,x) 

# here we can't use normal svm we need to add another dimension
```

```python
plt.scatter(y,x)
# inner circle
plt.scatter(y1,x1)
```

```python
np.vstack([y,x])
```

```python
# geting a transpose of outer circle
np.vstack([y,x]).T
```

```python
# for outer datapoint we label it as 0
import pandas as pd
df1 =pd.DataFrame(np.vstack([y,x]).T,columns=['X1','X2'])
df1['Y']=0
```

```python
# inner datapoints  1
df2 =pd.DataFrame(np.vstack([y1,x1]).T,columns=['X1','X2'])
df2['Y']=1
```

```python
# merge df 1 and df2 into df

df = pd.concat([df1,df2])
df
```

```python
df.head(5)
```

```python
df.tail()
```

```python
### Independent and Dependent features
X = df.iloc[:, :2]  
y = df.Y
```

```python
## Split the dataset into train and test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
y_train
```

```python
# We need to find components for the Polynomical Kernel
#X1,X2,X1_square,X2_square,X1*X2
df['X1_Square']= df['X1']**2
df['X2_Square']= df['X2']**2
df['X1*X2'] = (df['X1'] *df['X2'])
df.head()
```

```python
### Independent and Dependent features
X = df[['X1','X2','X1_Square','X2_Square','X1*X2']]
y = df['Y']
```

```python

X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size = 0.25, 
                                                    random_state = 0)
```

```python
# we convert the 2d into 3d 
# with the square
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
classifier = SVC(kernel="linear")
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
# here in ploy kernel it did all the calculation of inside we did not need to create it manually 
# create it internally

from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
classifier = SVC(kernel="poly")
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
# rbf ==> radial basis kernel
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
classifier = SVC(kernel="rbf")
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
classifier = SVC(kernel="sigmoid")
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
accuracy_score(y_test, y_pred)
```
