```python
df1 = pd.DataFrame({
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Marks': [25, 30, 28, 35]
    
})
df2 =pd.DataFrame({
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Course': ['Math', 'Science', 'English', 'History']})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2, on='Name')
# name factor is common in both dataframes. If we want to merge based on other column, we can specify it.
```
