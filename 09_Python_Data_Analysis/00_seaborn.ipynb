# Seaborn
    1. provides a high-level interface for drawing attractive and informative statistical graphics.
    2. It is built on top of Matplotlib and uses Pandas DataFrames.

```python
import seaborn as sns
```

```python
## Basic Plots in seaborn
tips = sns.load_dataset('tips')
# Tips dataset is a built-in dataset in seaborn
tips        
```

```python
## Create a scatter plot

import matplotlib.pyplot as plt
# data is entire data is all about tips
sns.scatterplot( x='total_bill', y='tip',  data=tips )
plt.title('Total Bill vs Tip')
```

```python
# line plot
sns.lineplot( x='size', y='tip', data=tips )

plt.title('Size vs Tip')
```

```python
# Catagorical scatter plot
# bar plot
sns.barplot( x='day', y='total_bill' ,data=tips )
plt.title('Total Bill vs day')

plt.show()
## Box plot
```

```python
# box plot for outlier detection

sns.boxplot( x='day', y='total_bill', data=tips )
plt.title('Total Bill vs Day')

plt.show()
## Pair plot
# as per persentile it will show outliers
```

```python
# Voilen plot

sns.violinplot( x='day', y='total_bill', data=tips )
plt.title('Total Bill vs Day')

plt.show()
## Heatmap

# show mean where white is high and black is low
```

```python
## Histogram 
# kde = kernel density estimation (Shape of graph)
sns.histplot( x='total_bill', data=tips , kde=True  , bins=10  )
plt.title('Total Bill Distribution')

plt.show()
## Facet grid
```

```python
## Histogram 
# kde = kernel density estimation (Shape of graph)
sns.histplot( x='total_bill', data=tips , kde=False  , bins=10  )
plt.title('Total Bill Distribution')

plt.show()
```

```python
## KDE plot

sns.kdeplot( x='total_bill', data=tips , shade=True  )
plt.title('Total Bill Distribution')

plt.show()
## Pair plot
```

```python
# Pairplot is used to visualize the relationship between multiple variables in a dataset. It is a great way to identify patterns and correlations between variables.

sns.pairplot( data=tips, hue='sex' )
plt.title('Pairplot with hue sex')

plt.show()
## Count plot
```

```python
sns.pairplot( data=tips )
# Need numerical data for count plot
# relationship between two variables
```

```python
# Heatmap
corr = tips[['total_bill','size','tip']].corr()
corr

# if total bill is increase size also increase and tip is increase size also increase by 0.5 unit and tip is increase by 0.6 unit
```

```python
sns.heatmap(corr, annot=True ,cmap="coolwarm")
```

```python
# Customizing 
import pandas as pd
sales_df = pd.read_csv('sales_data.csv')
```

```python
sales_df.head()
```

```python
# plot total sales by product category
plt.figure(figsize=(10,6))
sns.barplot( x='Product', y='Sales ($)', data=sales_df ,estimator= sum )
plt.title('Total Sales by Product Category')

plt.show()
# plot total sales by product category with hue of sales person
```

```python
# Total Sales by Region

plt.figure(figsize=(10,6))
sns.barplot( x='Region', y='Sales ($)', data=sales_df , estimator= sum)
plt.title('Total Sales by Region')

plt.show()
```
